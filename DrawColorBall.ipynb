```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def draw_balls(count):
    balls_in_bag = ['Red', 'Red', 'Red', 'Red', 'Red', 'Green', 'Green', 'Green', 'Blue']
    balls_outcome = []

    for _ in range(count):
        balls_outcome.append(np.random.choice(balls_in_bag))

    print('Drawn ball is RED: ', balls_outcome.count('Red'))
    print('Drawn ball is GREEN: ', balls_outcome.count('Green'))
    print('Drawn ball is BLUE: ', balls_outcome.count('Blue'))

    return balls_outcome


balls_outcome = draw_balls(10)
```

```python
plt.figure(figsize = (12, 6))

sns.countplot(x = balls_outcome, hue=balls_outcome, palette = {'Red' : 'red', 'Green':'green', 'Blue':'blue'})


plt.show()
```
