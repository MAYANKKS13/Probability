```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def roll_dice(count):
    dice_numbers = np.array([1, 2, 3, 4, 5, 6])
    dice_rolls = []
    for _ in range(count):
        dice_rolls.append(np.random.choice(dice_numbers))

    print('Probability of 1 comes on top: ', dice_rolls.count(1) / count)
    print('Probability of 2 comes on top: ', dice_rolls.count(2) / count)
    print('Probability of 3 comes on top: ', dice_rolls.count(3) / count)
    print('Probability of 4 comes on top: ', dice_rolls.count(4) / count)
    print('Probability of 5 comes on top: ', dice_rolls.count(5) / count)
    print('Probability of 6 comes on top: ', dice_rolls.count(6) / count)

    return dice_rolls

dice_rolls = roll_dice(100000)
```

```python
plt.figure(figsize = (12, 6))

sns.countplot(x = dice_rolls)

plt.show()
```

```python
from sympy.stats import P, Die

from sympy import Eq

X = Die('X', 6)

print('Probability of getting 1 at top: ', P(Eq(X, 1)))
print('Probability of getting 1 at top: ', P(Eq(X, 1)))
print('Probability of getting 1 at top: ', P(Eq(X, 1)))
print('Probability of getting 1 at top: ', P(Eq(X, 1)))
print('Probability of getting 1 at top: ', P(Eq(X, 1)))
print('Probability of getting 1 at top: ', P(Eq(X, 1)))
```
