```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def flip_coins(count):
    heads_tails = np.array([0, 1])
    coin_flips = []
    for _ in range(count):
        flip = np.random.choice(heads_tails)
        coin_flips.append(flip)

    print('Probability of HEADS: ', coin_flips.count(0) / count)
    print('Probability of TAILS: ', coin_flips.count(1) / count)

    return coin_flips

coin_flips = flip_coins(10000)
```

```python
plt.figure(figsize = (12, 6))

sns.countplot(x = coin_flips)

plt.show()
```
